```python
import umap
import pandas as pd 
from sklearn.decomposition import PCA
```

```python
df = pd.read_pickle('pkl/GI_40x256_120k30x_mar19_2025.pkl') 
df = df.sort_index()
df.head()
```

```python
df.shape
```

```python
features = df.iloc[:,:1024]
features.head()
```

```python
cluster_50 = pd.read_pickle('pkl/cluster_50.pkl')
cluster_50 = cluster_50.sort_index()
cluster_50.head()
```

```python
cluster_label = cluster_50['cluster'].values
cluster_label
```

```python
cluster_50.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
rescaled_features = scaler.fit_transform(features)
```

```python
rescaled_features
```

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Reduce to 2 components for 2D visualization
pca_2d = PCA(n_components=2)
components_2d = pca_2d.fit_transform(rescaled_features)

# 2. Create a new dataframe for plotting
# We pull the 'cancer_type' from your original df to color the points
plot_df = pd.DataFrame(data = components_2d, columns = ['PC1', 'PC2'])
plot_df['cancer_type'] = df['cancer_type'].values
plot_df['cluster_label'] = cluster_50['cluster'].values
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='PC1', y='PC2', hue='cancer_type', data=plot_df, palette='tab20', alpha=0.1, s=1, edgecolor=None)
plt.title('PCA of UNI Label Embeddings')
plt.xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]*100:.1f}% variance)')
plt.ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]*100:.1f}% variance)')
plt.savefig('png/pca_cancer.png') 
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='PC1', y='PC2', hue='cluster_label', data=plot_df, palette='nipy_spectral', alpha=0.1, s=1, edgecolor=None)
plt.title('PCA of UNI Embeddings')
plt.xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]*100:.1f}% variance)')
plt.ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]*100:.1f}% variance)')
plt.savefig('png/pca_2.png') 
plt.show()
```

```python
# 1. Reduce height (from 12 to 5 or 6 should be plenty for 3 columns)
fig_leg = plt.figure(figsize=(8, 6)) 

dummy_colors = sns.color_palette("nipy_spectral", n_colors=50)
handles = [plt.Line2D([0], [0], marker='o', color='w', 
                      markerfacecolor=c, markersize=10) for c in dummy_colors]
labels = [str(i) for i in range(50)]

# 2. Use loc='upper center' to snap it to the top near the title
# Added borderaxespad to give a tiny bit of breathing room from the title
plt.legend(handles, labels, title="Cluster Labels", 
           loc='upper center', ncol=3, fontsize='medium', 
           frameon=True, borderaxespad=1.5)

plt.axis('off')
plt.title("Cluster Color Key", fontsize=14, pad=20)

# 3. Use tight_layout to pull everything together
plt.tight_layout()
plt.show()
```

```python
import numpy as np 
pca_500 = PCA(n_components=500)
pc_500 = pca_500.fit_transform(rescaled_features)
pc_500 = pd.DataFrame(data = pc_500) 
pd.Series(np.cumsum(pca_500.explained_variance_ratio_)).plot()
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.savefig('png/pca_variance.png') 
plt.show()
```

```python
pca_df = pd.Series(np.cumsum(pca_500.explained_variance_ratio_))
pca_df
```

```python
pca_df.to_pickle('pkl/pca_variance.pkl') 
```

```python

reducer = umap.UMAP(
    n_neighbors=100,      # Increase for better global structure
    min_dist=0.1,        # 0.1 is standard; 0.0 creates tighter clusters
    metric='cosine',     # Use 'cosine' for high-dim embeddings/text/bio data
    low_memory=True,    # Keeps the RAM usage in check
    densmap=True, 
    verbose=True
)

embedding = reducer.fit_transform(pc_500)
```

```python
embedding
```

```python
labels = df['cancer_type'].values
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'embedding' is your (600000, 2) UMAP output
# and 'labels' is a list/array of your Cancer Types
plt.figure(figsize=(6, 4))
sns.scatterplot(
    x=embedding[:, 0], 
    y=embedding[:, 1], 
    hue=labels, 
    palette='tab20',
    s=0.1,         # Tiny points are key!
    alpha=0.3,     # Makes dense areas darker
    edgecolor=None
)

plt.title("UMAP Projection of UNI Embeddings", fontsize=10)
plt.xlabel('UMAP 1') 
plt.ylabel('UMAP 2') 
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', markerscale=10)
plt.tight_layout()
plt.show()
```
